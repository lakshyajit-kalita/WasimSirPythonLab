```python
#print("Hello")
#def AddTwoNumber(a, b):
 # sum = a + b
  #print("the summation of two numbers are :", sum)
#AddTwoNumber(5, 9)
numbers = 12345

def AddTwoNumber():
    n = numbers
    total = 0

    while n > 0:
        digit = n % 10      # get last digit
        total += digit      # add digit to total
        n = n // 10         # remove last digit

    return total

print(AddTwoNumber())
```

```python
# 03_Reverse a number (no string).

n = 1234
def reverseNum():
  num = n
  rev = 0

  while num > 0 :
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

  return rev
print(reverseNum())
```

```python
#Check palindrome number (no string).
n = int(input("Enter a number: "))

original = n
rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

if rev == original:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

```python
#Check Armstrong number (general n-digit).
n = int(input("Enter a number: "))

temp = n
digits = 0
sum = 0

# count digits
while temp > 0:
    digits += 1
    temp //= 10

temp = n

while temp > 0:
    digit = temp % 10
    sum += digit ** digits
    temp //= 10

if sum == n:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

```python
#Check Perfect number.
n = int(input("Enter a number: "))

sum = 0

for i in range(1, n):
    if n % i == 0:
        sum += i

if sum == n:
    print("Perfect number")
else:
    print("Not a Perfect number")
```

```python
#Check Strong number (sum of factorial of digits).
n = int(input("Enter a number: "))

temp = n
sum = 0

while temp > 0:
    digit = temp % 10
    fact = 1

    for i in range(1, digit + 1):
        fact *= i

    sum += fact
    temp //= 10

if sum == n:
    print("Strong number")
else:
    print("Not a Strong number")
```

```python
#Check Harshad/Niven number.
n = int(input("Enter a number: "))

temp = n
sum = 0

while temp > 0:
    sum += temp % 10
    temp //= 10

if n % sum == 0:
    print("Harshad (Niven) number")
else:
    print("Not a Harshad number")
```

```python
#Check Automorphic number (ends with its square).
n = int(input("Enter a number: "))

square = n * n
temp = n

while temp > 0:
    if temp % 10 != square % 10:
        print("Not an Automorphic number")
        break
    temp //= 10
    square //= 10
else:
    print("Automorphic number")
```

```python
#10.Check Neon number (sum digits of square equals number)
n = int(input("Enter a number: "))

square = n * n
sum = 0

while square > 0:
    sum += square % 10
    square //= 10

if sum == n:
    print("Neon number")
else:
    print("Not a Neon number")
```
