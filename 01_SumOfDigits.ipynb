```python
#print("Hello")
#def AddTwoNumber(a, b):
 # sum = a + b
  #print("the summation of two numbers are :", sum)
#AddTwoNumber(5, 9)
numbers = 12345

def AddTwoNumber():
    n = numbers
    total = 0

    while n > 0:
        digit = n % 10      # get last digit
        total += digit      # add digit to total
        n = n // 10         # remove last digit

    return total

print(AddTwoNumber())
```

```python
# 03_Reverse a number (no string).

n = 1234
def reverseNum():
  num = n
  rev = 0

  while num > 0 :
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

  return rev
print(reverseNum())
```
